```python
train=df['Southern Osc'][:400]
test=df['Southern Osc'][-56:]
plt.figure(figsize=(20,20))
plt.ylabel('Oscillations')
plt.plot(train)
plt.plot(test)
```

```python
roll_mean=train.rolling(window=10).mean()
roll_std=train.rolling(window=10).std()
plt.figure(figsize=(20,20))
plt.plot(train)
plt.plot(roll_mean,color='r')
plt.plot(roll_std,color='y')
```

```python
from statsmodels.tsa.stattools import adfuller
adf_test=adfuller(train)
adf_test[1]
```

```python
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
ACF=plot_acf(train)
PACF=plot_pacf(train)
ACF,PACF
```

```python
from statsmodels.tsa.arima.model import ARIMA
arima=ARIMA(train,order=(5,0,10))
model=arima.fit()
```

```python
pred=model.forecast(len(test))
plt.figure(figsize=(20,20))
plt.plot(train)
plt.plot(test)
plt.plot(pred,color='g')
```
